# 01 — Blob Storage smoke test

Verifies that `BlobService` can:
1. Connect to the configured container
2. Upload a small file
3. List blobs
4. Download the file
5. Delete the file

**Prereq:** `.env` is filled in with `AZURE_STORAGE_ACCOUNT_NAME`, `AZURE_BLOB_CONTAINER_NAME`, and either `AZURE_BLOB_SAS_TOKEN` or a working managed identity.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
from src.blob_client import BlobService

blob = BlobService()
blob.ensure_container()
print('Container:', blob.container)
```

```python
# Upload
name = 'docmind-test/hello.txt'
url = blob.upload(name, b'hello docmind', content_type='text/plain')
print('Uploaded:', url)
```

```python
# List
for n in list(blob.list_blobs(prefix='docmind-test/'))[:10]:
    print(n)
```

```python
# Download
data = blob.download(name)
print('Downloaded:', data)
```

```python
# Cleanup
blob.delete(name)
print('Deleted ✔')
```
